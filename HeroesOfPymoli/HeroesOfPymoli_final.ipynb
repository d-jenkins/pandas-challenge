## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_list = purchase_df['Age'].tolist()
age_list.sort()
age_list
```

```python
age_bin = [0, 9, 14, 19, 24, 29, 34, 39, 45]
# Create the names for the five bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
age_groups = purchase_df["Age Group"] = pd.cut(purchase_df["Age"], age_bin, labels=group_names, include_lowest=True).value_counts()
age_groups.head()
```

```python
# Creating a group based off of the bins
grouped_by_age = purchase_df.groupby("Age Group")
grouped_by_age.min()
```
